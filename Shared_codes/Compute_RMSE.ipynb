```python
import glob
import pandas as pd
import numpy as np

# Get all EONR CSV files
file_list = glob.glob('/Users/mmousavi2/Dropbox/Causal_climate/My_own_Shared/Takashi_code/Rlearner/Codes/Test/EONR_*.csv')

# Initialize lists to store the averages
avg_true = []
avg_pred = []

# Process each file to compute averages
for file in file_list:
    df = pd.read_csv(file)
    avg_true.append(df['true'].mean())
    avg_pred.append(df['pred'].mean())

# Convert lists to numpy arrays for vectorized operations
avg_true = np.array(avg_true)
avg_pred = np.array(avg_pred)

# Compute RMSE
rmse = np.sqrt(np.mean((avg_pred - avg_true) ** 2))
print("RMSE:", rmse)
```

```python
import os
import glob
import pandas as pd
import numpy as np

# Folder names and display titles
folders_titles = [
    ('Half_RANN_1_field', 'Half R ANN 1 field'),
    ('Half_RANN_5_fields', 'Half R ANN 5 fields'),
    ('Half_RANN_10fields', 'Half R ANN 10 fields'),
    ('RANN_Pytorch_outcome_one_field', 'R ANN 1 field'),
    ('RANN_Pytorch_outcome_five_field', 'R ANN 5 fields'),
    ('RANN_Pytorch_outcome_ten_fields', 'R ANN 10 fields'),
    ('Simple_ANN_1field', 'Simple ANN 1 field'),
    ('Simple_ANN_5fields', 'Simple ANN 5 fields'),
    ('Simple_ANN_10fields', 'Simple ANN 10 fields'),
    ('Half_RRF_1_field', 'Half R RF 1 field'),
    ('Half_RRF_5_fields', 'Half R RF 5 fields'),
    ('Half_RRF_10_fields', 'Half R RF 10 fields'),
    ('CF_one_Field', 'CF 1 field'),
    ('CF_Five_fields', 'CF 5 field'),
    ('CF_ten_Fields', 'CF 10 field')
]

base_path = '/Users/mmousavi2/Dropbox/Causal_climate/My_own_Shared/Results_so_not_confuded_git/Results'

# List to collect RMSE results
rmse_summary = []

for folder, title in folders_titles:
    file_list = glob.glob(os.path.join(base_path, folder, 'EONR_*.csv'))

    avg_true = []
    avg_pred = []

    for file in file_list:
        try:
            df = pd.read_csv(file)
            avg_true.append(df['true'].mean())
            avg_pred.append(df['pred'].mean())
        except Exception as e:
            print(f"Error in {file}: {e}")

    # Compute RMSE if data exists
    if avg_true and avg_pred:
        avg_true = np.array(avg_true)
        avg_pred = np.array(avg_pred)
        rmse = np.sqrt(np.mean((avg_pred - avg_true) ** 2))
    else:
        rmse = np.nan  # fallback if data is missing

    rmse_summary.append({'Model': title, ' RMSE': rmse})

# Convert to DataFrame
rmse_df = pd.DataFrame(rmse_summary)

# Display
rmse_df
```
