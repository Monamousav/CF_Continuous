```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Folder names and titles
folders_titles = [
    ('Half_RANN_1_field', 'Half R ANN 1 field'),
    ('Half_RANN_5_fields', 'Half R ANN 5 fields'),
    ('Half_RANN_10fields', 'Half R ANN 10 fields'),
    ('RANN_Pytorch_outcome_one_field', 'R ANN 1 field'),
    ('RANN_Pytorch_outcome_five_field', 'R ANN 5 fields'),
    ('RANN_Pytorch_outcome_ten_fields', 'R ANN 10 fields'),
    ('Simple_ANN_1field', 'Simple ANN 1 field'),
    ('Simple_ANN_5fields', 'Simple ANN 5 fields'),
    ('Simple_ANN_10fields', 'Simple ANN 10 fields'),
    ('Half_RRF_1_field', 'Half R RF 1 field'),
    ('Half_RRF_5_fields', 'Half R RF 5 fields'),
    ('Half_RRF_10_fields', 'Half R RF 10 fields'),
    ('CF_one_Field', 'CF 1 field'),
    ('CF_Five_fields', 'CF 5 field'),
    ('CF_ten_Fields', 'CF 10 field'),
]

base_path = '/Users/mmousavi2/Dropbox/Causal_climate/My_own_Shared/Results_so_not_confuded_git/Results'

# Prepare RMSE list for all folders
all_rmse_lists = []

for folder_name, _ in folders_titles:
    rmse_list = []
    folder_path = os.path.join(base_path, folder_name)
    
    for i in range(1, 501):
        file_name = f"EONR_{i}.csv"
        file_path = os.path.join(folder_path, file_name)
        try:
            df = pd.read_csv(file_path)
            df.columns = df.columns.str.strip()

            if {'pred', 'true'}.issubset(df.columns):
                rmse = np.sqrt(np.mean((df['pred'] - df['true']) ** 2))
                rmse_list.append(rmse)
        except Exception as e:
            print(f"Error reading {file_name} in {folder_name}: {e}")
    
    all_rmse_lists.append(rmse_list)

# Plot all histograms in subplots
fig, axs = plt.subplots(5, 3, figsize=(10, 18))  # 5 rows, 3 columns
axs = axs.flatten()

for i, (rmse_list, (folder_name, title)) in enumerate(zip(all_rmse_lists, folders_titles)):
    ax = axs[i]
    ax.hist(rmse_list, bins=30, color='skyblue', edgecolor='black')
    ax.set_title(title, fontsize=11, family='serif')
    ax.set_xlim(0, 120)
    ax.set_ylim(0, 100)
    ax.set_xticks([0, 50, 100])
    ax.set_yticks([0, 20, 40, 60, 80, 100])
    ax.tick_params(labelsize=10)
    for spine in ax.spines.values():
        spine.set_linewidth(1.2)
        spine.set_color('black')

# Add a single shared x-label
fig.text(0.5, 0.04, r'RMSE (N kg ha$^{-1}$)', ha='center', fontsize=13)
# Adjust spacing
plt.tight_layout(rect=[0, 0.05, 1, 1])
plt.show()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

folder_path = '/Users/monamousavi/Dropbox/Causal_climate/My_own_Shared/Causal-ML-continuous-treatment/Code/CF/Test_cf'

rmse_list = []

for i in range(1, 501):
    file_name = f"EONR_{i}.csv"
    file_path = os.path.join(folder_path, file_name)
    
    try:
        df = pd.read_csv(file_path)
        df.columns = df.columns.str.strip()  # Remove extra spaces from column names

        if i % 50 == 0:
            print(f"Processing {file_name}")
        
        if {'pred', 'true'}.issubset(df.columns):
            rmse = np.sqrt(np.mean((df['pred'] - df['true']) ** 2))
            rmse_list.append(rmse)
        else:
            print(f"Columns missing in {file_name}")

    except Exception as e:
        print(f"Error reading {file_name}: {e}")

import matplotlib.pyplot as plt

# Plotting
if rmse_list:
    plt.figure(figsize=(5, 7))  # Tall and narrow

    plt.hist(rmse_list, bins=30, color='skyblue', edgecolor='black')
    
    # Labels and title
    plt.xlabel(r'RMSE (N kg ha$^{-1}$)', fontsize=12)
    plt.ylabel('')
    plt.title('CF', fontsize=14, family='serif')

    # Tick marks
    plt.xticks([0, 50, 100], fontsize=12)
    plt.yticks([0, 20, 40, 60, 80, 100], fontsize=12)

    # Axis limits
    plt.xlim(0, 120)
    plt.ylim(0, 100)

    # Ensure all 4 spines (borders) are visible and bold
    ax = plt.gca()
    for spine in ax.spines.values():
        spine.set_linewidth(1.5)
        spine.set_color('black')
        spine.set_visible(True)

    plt.tight_layout()
    plt.show()
else:
    print("No RMSE values were calculated.")



```
