```python
import glob
import pandas as pd
import matplotlib.pyplot as plt

# Get all EONR CSV files
file_list = glob.glob('./Pytorch_outcome_five_field/EONR_*.csv')

# Initialize lists to store the averages
avg_true = []
avg_pred = []
sim_ids = []  # optional: to keep track of simulation id if needed

# Process each file to compute averages
for file in file_list:
    df = pd.read_csv(file)
    avg_true.append(df['true'].mean())
    avg_pred.append(df['pred'].mean())
    
    # Optional: extract simulation id from filename
    sim_id = file.split('_')[-1].replace('.csv', '')
    sim_ids.append(sim_id)

# Create a single scatter plot for the averages
plt.figure(figsize=(8, 8))
plt.scatter(avg_true, avg_pred, alpha=0.7)
plt.xlabel('Average True EONR')
plt.ylabel('Average Predicted EONR')
plt.title('Average True vs. Predicted EONR Across Simulations')

# Plot the y = x reference line
min_val = min(min(avg_true), min(avg_pred))
max_val = max(max(avg_true), max(avg_pred))
plt.plot([min_val, max_val], [min_val, max_val], 'r--', label='y = x')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import glob
import pandas as pd
import numpy as np

# Get all EONR CSV files
file_list = glob.glob('./Pytorch_outcome_one_field/EONR_*.csv')

# Initialize lists to store the averages
avg_true = []
avg_pred = []

# Process each file to compute averages
for file in file_list:
    df = pd.read_csv(file)
    avg_true.append(df['true'].mean())
    avg_pred.append(df['pred'].mean())

# Convert lists to numpy arrays for vectorized operations
avg_true = np.array(avg_true)
avg_pred = np.array(avg_pred)

# Compute RMSE
rmse = np.sqrt(np.mean((avg_pred - avg_true) ** 2))
print("RMSE:", rmse)
```

```python
import glob
import math
import pandas as pd
import matplotlib.pyplot as plt
import os

# Your folders
folders = [
    'Res_second_time_10_fields',
    'Results/Pytorch_outcome_five_field',
    'Results/Pytorch_outcome_one_field'
]
folder_labels = ['10 Fields', '5 Fields', '1 Field']

# You want the first 10 EONR files
eonr_ids = range(1, 11)  # EONR_1 to EONR_10

rows = len(eonr_ids)
cols = len(folders)

fig, axes = plt.subplots(rows, cols, figsize=(cols*5, rows*5), squeeze=False)

for row_idx, eonr_id in enumerate(eonr_ids):
    for col_idx, folder in enumerate(folders):
        filename = f'./{folder}/EONR_{eonr_id}.csv'
        if not os.path.exists(filename):
            print(f"File not found: {filename}")
            continue
        
        df = pd.read_csv(filename)  # Read full CSV
        
        ax = axes[row_idx, col_idx]
        ax.scatter(df['true'], df['pred'], alpha=0.7)
        ax.set_xlabel('True EONR')
        ax.set_ylabel('Predicted EONR')

        ax.set_title(f"{folder_labels[col_idx]} - sim_{eonr_id}")

        # Plot y = x line
        min_val = min(df['true'].min(), df['pred'].min())
        max_val = max(df['true'].max(), df['pred'].max())
        ax.plot([min_val, max_val], [min_val, max_val], 'r--')

plt.tight_layout()
plt.show()
```

```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import os

# Folders to process
folders = {
    '10 Fields': 'Pytorch_outcome_ten_fields',
    '5 Fields': 'Pytorch_outcome_five_field',
    '1 Field': 'Pytorch_outcome_one_field'
}

# Dictionary to store results
results = {}

# Read and process each folder
for label, folder in folders.items():
    avg_trues = []
    avg_preds = []

    file_list = sorted(glob.glob(f'./{folder}/EONR_*.csv'))  # Ensure ordering
    for file in file_list:
        df = pd.read_csv(file)
        avg_trues.append(df['true'].mean())
        avg_preds.append(df['pred'].mean())

    # Store in dictionary
    results[label] = pd.DataFrame({
        'avg_true': avg_trues,
        'avg_pred': avg_preds
    })

# Plotting
fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharex=True, sharey=True)

for idx, (label, df) in enumerate(results.items()):
    ax = axes[idx]
    ax.scatter(df['avg_true'], df['avg_pred'], alpha=0.6)
    ax.plot([df['avg_true'].min(), df['avg_true'].max()],
            [df['avg_true'].min(), df['avg_true'].max()], 'r--')
    ax.set_title(f'{label}')
    ax.set_xlabel('Average True EONR')
    if idx == 0:
        ax.set_ylabel('Average Predicted EONR')

plt.tight_layout()
plt.show()
```
